```python
def calcular_estadisticos_historicos(df):
    """
    Calcula, para cada fila, las estadísticas del estudiante hasta el ciclo anterior.
    Incluye promedio ponderado, notas, asistencias y créditos aprobados.
    """
    df = df.copy()
    resultados = []

    # Agrupar por estudiante
    for cod, grupo in df.groupby("COD_ALUMNO", sort=False):
        historial = []
        for idx, row in grupo.iterrows():
            # Filtrar cursos anteriores
            anteriores = grupo[grupo["PER_MATRICULA"] < row["PER_MATRICULA"]]
            
            if len(anteriores) == 0:
                resultados.append({
                    "PROM_POND_HIST": 0,
                    "NOTA_MAX_HIST": 0,
                    "NOTA_MIN_HIST": 0,
                    "NOTA_MEDIAN_HIST": 0,
                    "NOTA_Q1_HIST": 0,
                    "NOTA_Q3_HIST": 0,
                    "ASIST_PROM_HIST": 0,
                    "CRED_APROB_HIST": 0
                })
                continue
            
            # Calcular métricas
            notas = anteriores["NOTA"]
            asist = anteriores["PRCTJE_INASISTENCIA"]
            cred = anteriores.loc[anteriores["APROBO"] == "Si", "CREDITOS"]
            
            ponderado = np.average(anteriores["NOTA"], weights=anteriores["CREDITOS"])
            q1, q3 = notas.quantile(0.25), notas.quantile(0.75)
            
            resultados.append({
                "PROM_POND_HIST": ponderado,
                "NOTA_MAX_HIST": notas.max(),
                "NOTA_MIN_HIST": notas.min(),
                "NOTA_MEDIAN_HIST": notas.median(),
                "NOTA_Q1_HIST": q1,
                "NOTA_Q3_HIST": q3,
                "ASIST_PROM_HIST": asist.mean(),
                "CRED_APROB_HIST": cred.sum()
            })
    
    hist_df = pd.DataFrame(resultados)
    df = pd.concat([df.reset_index(drop=True), hist_df], axis=1)
    return df
```

```python
def calcular_estadisticos_cluster(df):
    """
    Para cada fila, calcula estadísticas históricas del estudiante dentro de su cluster.
    Incluye promedio, cursos aprobados y desaprobados.
    """
    df = df.copy()
    resultados = []

    for (alumno, cluster), grupo in df.groupby(["COD_ALUMNO", "CLUSTER_CURSO"], sort=False):
        for idx, row in grupo.iterrows():
            anteriores = grupo[grupo["PER_MATRICULA"] < row["PER_MATRICULA"]]
            
            if len(anteriores) == 0:
                resultados.append({
                    "NOTA_PROM_CLUSTER_HIST": 0,
                    "CURSOS_APRO_CLUSTER_HIST": 0,
                    "CURSOS_DESAPRO_CLUSTER_HIST": 0
                })
                continue
            
            notas = anteriores["NOTA"]
            apro = (anteriores["APROBO"] == "Si").sum()
            desapro = (anteriores["APROBO"] == "No").sum()
            
            resultados.append({
                "NOTA_PROM_CLUSTER_HIST": notas.mean(),
                "CURSOS_APRO_CLUSTER_HIST": apro,
                "CURSOS_DESAPRO_CLUSTER_HIST": desapro
            })
    
    cluster_df = pd.DataFrame(resultados)
    df = pd.concat([df.reset_index(drop=True), cluster_df], axis=1)
    return df
```

```python
def generar_estadisticos_completos(df):
    df = calcular_estadisticos_historicos(df)
    df = calcular_estadisticos_cluster(df)
    return df
```

```python
# Supongamos que ya cargaste tu DataFrame completo_df_cargado
df_resultado = generar_estadisticos_completos(completo_df_cargado)

# Vista previa
print(df_resultado.head())
```
