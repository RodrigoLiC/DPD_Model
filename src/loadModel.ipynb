```python
import pickle
import pandas as pd

def predecir_con_pickle(pkl_path, columnas, tupla_datos):
    """
    Carga un modelo desde un archivo .pkl (guardado con pickle)
    y realiza una predicción sobre una tupla de características.

    Parámetros:
    -----------
    pkl_path : str
        Ruta al archivo .pkl del modelo entrenado.
    columnas : list[str]
        Nombres de las columnas en el mismo orden usado para entrenar el modelo.
    tupla_datos : tuple o list
        Valores de entrada (en el mismo orden que 'columnas').

    Retorna:
    --------
    float
        Predicción del modelo (por ejemplo, nota estimada).
    """

    # Cargar modelo desde archivo pickle
    with open(pkl_path, "rb") as f:
        model = pickle.load(f)

    # Convertir la tupla a un DataFrame con las columnas esperadas
    X_nuevo = pd.DataFrame([tupla_datos], columns=columnas)

    # Hacer predicción
    pred = model.predict(X_nuevo)[0]

    print(f"📘 Predicción del modelo cargado desde {pkl_path}: {pred:.4f}")
    return pred
```
